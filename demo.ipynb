```python
# Let's set the initial conditions for two pendula via the Setup

s1=RottSetup(g0=0.,gdot0=0.0,adot0=1000.)
s2=RottSetup(g0=0.,gdot0=0.0,adot0=1000.01) # at t=0 the second pendulum left arm has a slightly higher angular velocity adot0

pp1=RottProperties(s1)    
pp2=RottProperties(s2)
```

```python
t1=RottTrajectoryInPhasespace(s1,pp1)
t1.integrate(int_scheme=0) # integrate the pendulums dynamic for 60s

t2=RottTrajectoryInPhasespace(s2,pp2)
t2.integrate(int_scheme=0) # integrate the pendulums dynamic for 60s
```

Now let's inspect how the the two pendulas motion evolve with time:

```python
f,ax=plt.subplots()
t1.plot(f=f,ax=ax)
t2.plot(f=f,ax=ax)

ax.set_xlim([0,3])
ax.set_xlabel("t in s")
ax.set_ylabel(r"$\alpha$ and $\gamma$ in rad / s")
```

Can you see how the respective angular speeds start to differ after 1.5 s? Reducing the difference in initial conditions will have a logarithmically small impact.
